```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import us
```

```python
elections = {}
```

```python
with open("res/house_general_election_2016.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if (row['state'] + row['fips']) not in elections:
            elections[(row['state']+row['fips'])] = {}
        #print(elections)
        if row['vote_pct'] == '':
            row['vote_pct'] = '100'
        
        if row['individual_party'] in elections[str(row['state']+row['fips'])]:
            elections[str(row['state']+row['fips'])][row['individual_party']] += float(row['vote_pct'])
        else:
            elections[str(row['state']+row['fips'])][row['individual_party']] = float(row['vote_pct'])
```

```python
elections
```

```python
for race in elections:
    #print(race)
    for key in ['republican', 'democrat', 'independent_or_other']:
        #print(key)
        if key not in elections[race]:
            elections[race][key] = 0
        elif elections[race][key] == '':
            elections[race][key] = 100
        else:
            elections[race][key] = float(elections[race][key])
```

```python
ea = np.zeros((435, 6))
e17 = {}
e17['Kansas4'] = {'republican':63505,'democrat':55310,'independent_or_other':2082}
e17['Georgia6'] = {'republican':98192,'democrat':94201,'independent_or_other':176}
e17['Montana1'] = {'republican':189473,'democrat':166483,'independent_or_other':21509}
e17['California34'] = {'republican':1089,'democrat':25581,'independent_or_other':1864}
for e in e17:
    e17[e]['margin'] = 100*(e17[e]['democrat']-e17[e]['republican'])/(e17[e]['republican']+e17[e]['democrat']+e17[e]['independent_or_other'])
el = []
for race in elections:
    #print(elections[race])
    r = [0, 0, 0, 0, 0]
    r[0] = elections[race]['republican']
    r[1] = elections[race]['democrat']
    r[2] = elections[race]['independent_or_other']
    r[3] = elections[race]['democrat'] - elections[race]['republican']
    if race in e17:
        r[4] = e17[race]['margin']#*100/(e17[race]['republican']+e17[race]['democrat']+e17[race]['independent_or_other'])
    #if 300000 not in r:
    el.append(r)
    
print(len(el))
print(np.asarray(el))
```

```python
comps = []
for race in el:
    #print(race)
    if race[4] != 0:
        comps.append(race)
        
print(np.asarray(comps))
```

```python
from sklearn import linear_model

comps = np.asarray(comps)
print(comps)
print(comps[:,3])
print(comps[:,4])
reg = linear_model.LinearRegression(fit_intercept=False)
reg.fit(comps[:,3].reshape(-1,1), comps[:,4])
print(reg)


x_test = np.asarray(el)
#print(x_test)
y = reg.predict(x_test[:,3].reshape(-1,1))

results = np.c_[x_test[:,3], y]
print(results)
print(sum(results[:,4]))
```

```python
#print(results)
changes = 0
old = 0
wins = 0
for r in results:
    changes += r[1]/435
    old += r[0]/435
    if r[1] > 48:
        wins += 1
print(changes)
print(old)
print(wins*100/435)
```
